```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_acc(log):
    acc = []
    with open(log, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if line[:5] == 'epoch':
                if len(line.split(',')) == 6:
                    acc.append(100*float(line.split(',')[-1][9:]))
                else:
                    acc.append(100*float(line.split(',')[-2][9:]))
    return acc
```

```python
log1 = './images/test_training/runs870/logs.log'
log2 = './images/test_training/runs868/logs.log'
log3 = './images/test_training/runs871/logs.log'

acc1 = read_acc(log1)
acc2 = read_acc(log2)
acc3 = read_acc(log3)
plt.plot(acc1, label='Init-1', color='b')
plt.plot(acc3, label='Init-2', color='r')
plt.plot(acc2[:200], label='Init-3', color='g')
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Accuracy(%)', fontsize=14)
plt.legend(fontsize=11)
plt.grid(linestyle=':')
```
