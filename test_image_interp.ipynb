```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = np.array(Image.open('test_BICUBIC.jpg'))
img2 = np.array(Image.open('test_BILINEAR.jpg'))
print(img1.shape, img2.shape)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
import cv2

image = cv2.imread('test_BICUBIC.jpg', cv2.IMREAD_GRAYSCALE)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
plt.imshow(gradient_magnitude)
```

```python
image = cv2.imread('test_BILINEAR.jpg', cv2.IMREAD_GRAYSCALE)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
plt.imshow(gradient_magnitude)
```
