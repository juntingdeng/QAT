```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
img1 = np.array(Image.open('test_BICUBIC.jpg'))
img2 = np.array(Image.open('test_BILINEAR.jpg'))
print(img1.shape, img2.shape)
```

```python
var1 = np.std(img1)
fft1 = np.fft.fft2(img1)
fftshift1 = np.fft.fftshift(fft1)
fftshift1_mag = np.abs(fftshift1)
fftshift1_mag = fftshift1_mag/fftshift1_mag.max()
print(var1)
plt.imshow(np.log(fftshift1_mag+1), cmap='gray')
```

```python
var2 = np.std(img2)
fft2 = np.fft.fft2(img2)
fftshift2 = np.fft.fftshift(fft2)
fftshift2_mag = np.abs(fftshift2)
fftshift2_mag = fftshift2_mag/fftshift2_mag.max()
print(var2)
plt.imshow(np.log(fftshift2_mag+1), cmap='gray')
```

```python
import cv2

image = cv2.imread('test_BICUBIC.jpg', cv2.IMREAD_GRAYSCALE)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
plt.imshow(gradient_magnitude)
```
